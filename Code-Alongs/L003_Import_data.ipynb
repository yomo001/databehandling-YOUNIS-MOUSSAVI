# Dataframe from file
Pandas supports loading data from multiple file formats, including: excel, csv, json, xml
- pd.read_csv/json/etc..


```python
import pandas as pd
```

```python
pd.read_csv("..\\Data\\students.csv") # Default: Första rad blir header
pd.read_csv("..\\Data\\students.csv", header = None) # Första rad blir inte header om man sätter header parameter = None
pd.read_csv("..\\Data\\students.csv", header = 1) # Om man vill skippa första raden och att header istället börjar på rad med index 1 (rad 2).
```

```python
# Import av csv fil
students = pd.read_csv("..\\Data\\students.csv") 
students.head(1) # får upp första raden (inklusive header), default är 5 första raderna
```

```python
# Import av json fil
autos = pd.read_json("..\\Data\\autos.json")
autos.head()   # får upp 5 första raderna
```

```python
# Import av excel fil från statistiska centralbyrån

scb = pd.read_excel("..\\Data\\befolkningsstatistik.xlsx", skiprows = 3)  
    # Man kan även ange sheetname och annat, se AIP för alla parametrar)
    # I denna fil finns det text innan, och tabellen börjar på annan rad. Då kan man använda skiprows eller header
scb.head()        # Ser första 5 raderna, och vilka rows som kan skippas.
```

```python
# Tillbaka till autos json filen

autos.head(10)                # Ser de första 10 raderna i filen
autos.tail()                  # Får upp 5 sista raderna i filen
autos.sample(5)               # Får fram fem olika slumpmässiga rader
autos                         # Får fram hela listan
```

```python
# Vid många kolumner och rader så visas några kolumner/rader, sedan kommer tre små punkter och bryter listan och därefter resten. Detta görs p.g.a. inställning i panda
# som kan ändras:

pd.set_option("display.max_columns", None)          # Här ser man samtliga, man ändrar alltså från default som är 20 till ingen maxgräns
pd.set_option("display.max_rows", None)             # Här ser man samtliga, man ändrar alltså från default som är 20 till ingen maxgräns

pd.set_option("display.max_columns", 6)       # Visar alla kolumner, men enbart första tre och sista tre raderna.
pd.set_option("display.max_rows", 6)          # Visar alla rader, men enbart första tre och sista tre raderna.
```

```python
autos.shape  # detta visar antal rader och kolumner i listan/filen

rows, columns = autos.shape         # Unpacking, men delar tuplen från shape i två variabler

print(f"Dataframe has {rows} and {columns} columns")
```

```python
# Går även att ta reda på antal kolumner och rader på detta sätt: 
rows = len(autos)
columns = len(autos.columns)

print(f"Dataframe has {rows} and {columns} columns")

```

```python
autos.info() # antal kolumner, vilka datatyper, vilka saknar data, minnet som tabellen tar, m.m.
```

### Convert to common formats
När vi vill spara eller konvertera gör vi det direkt i dataframes, förut enbart läst dataframes men här sparar ned i nytt format
Då väljer man "to" och sedan något, enligt:
filnamn.to_format

```python
# Copy dataframe to clipboard (then use Ctrl + V)
autos.to_clipboard()        # Här kopierar man dataframe till clipboard.
                            # Man kan sedan klistra in i t.ex. Excel, google docs osv. Smidigt att kunna göra så
```

```python
# Konvertera datafram till viss datatyp

autos.to_csv("..\\Data\\autos.csv", index = None) # Man anger filnamn och format att omvandla till, sedan vart filen ska sparas och heta
                                                  # Man får med index i panda dataframes 
                                                  # i utskrift men man vill inte ha med det i filen och använder index = None

csv_string = autos.to_csv()                       # Här refererar man variabel csv_string till den nya filen i csv format

print(csv_string[:1000])                          # print för att få ut som rader, och sedan första 1000 tecken.
                                                  # om man skriver csv_string direkt så fås allt på en och samma rad
```

```python
autos.to_json("..\\Data\\autos.json", orient = "records")     # Orient records används för att få samma utseende som först. Annars i kolumner
                                                              # Annars är default att det görs kolumnvis
                                                              # Räcker att känna till här att man kan ta det kolumn eller radvis
                                                              # När man skriver över till json

json_string = autos.to_json(indent = 4, orient = "records")  # Här refererar variabeln till detta nya omvandlade format
print(json_string[:200])                                     # Skriver ut första 200 tecken 
                                                                                                                            
```

```python
my_dict = autos.to_dict(orient = "records")    # man tar en Panda dataframe och gör om till dictionary
                                               # jämför med första exempel där man istället tog dictionary och gjorde till panda dataframe
my_dict[0]                                     # Första index kan man ta ut eftersom det är lista med dictionaries nu
```
