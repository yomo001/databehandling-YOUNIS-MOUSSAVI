## Pandas time series
A series or dataframe using a datetime index instead of a range index.

Used for storing events/data that fits on a timeline
- Weather data
- Temperature readings
- Heart rate monitoring(EKG)
- Quaterly sales
- Stock prices

```python
import pandas as pd
import numpy as np
```

### Creating a datetimeindex

Use date_range() with 3 of the 4 parameters:
- start
- end
- periods
- freq

```python
datetimeindex = pd.date_range(start = '2018-01-01', end = '2018-02-10', periods= 10)

type(datetimeindex)
```

```python
datetimeindex 
```

```python
datetimeindex = pd.date_range(start = '2018-01-01', end = '2018-02-10', freq= 'D')   # se fler sådana offset aliases på pandas dokumnetation, men flera är : freq= 'D' is default, freq= 'B' is business days, freq= 'W' is week, freq= 'M' is month end, freq= 'MS' is month start, freq= 'A' is year end, freq= 'AS' is year start

datetimeindex
```

```python
datetimeindex = pd.date_range(start = '2018-01-01', freq='w', periods= 10)
datetimeindex
```

```python
datetimeindex = pd.date_range(start = '2018-01-01', freq='2D', periods= 10)    # se fler sådana offset aliases på pandas dokumnetation, freq='2D' is every second day, freq='2D5H' is every second day and 5 hours.
datetimeindex
```

```python
datetimeindex = pd.date_range(start = '2018-01-01', freq='D', periods= 10)    
```

```python
df = pd.DataFrame(
    dict(
        n = range(len(datetimeindex)),
        rand = np.random.rand(len(datetimeindex))
    ), index = datetimeindex           # Antingen direkt skriver in index i lista eller väljer datetimeindex som index för tid
)

df
```

### Resampling is the practice of creating new samples with a lower or higher frequency than the original data

#### Downsampling
When the sample frequency is lower than the original we can aggregate values from the original data.

```python
df
```

```python
from helpers import hdisplay

resample_method = "Y"            # se fler sådana reset alias. resample är en metod som kan användas på en dataframe, och den tar en frekvens som argument, och den kan vara 'M' för månad, 'W' för vecka, 'D' för dag, 'Y' för år, 'MS' för månad start, 'AS' för år start, 'B' för business day, 'H' för hour, 'T' för minute, 'S' för second, 'L' för millisecond, 'U' för microsecond, 'N' för nanosecond.

hdisplay([
df.head(10),
df.resample(resample_method).agg({"n":"max", "rand":"sum"}).head(10)],
["Original", f"Resampled using '{resample_method}'"]
)

```

### Upsampling
When the sample frequency is higher than the original, we need to fill in the gaps. This is called upsampling.
- ffill
- bfill
- nearest
- interpolate
- fillna

```python
resample_method = "6H"            # se fler sådana reset alias. resample är en metod som kan användas på en dataframe, och den tar en frekvens som argument, och den kan vara 'M' för månad, 'W' för vecka, 'D' för dag, 'Y' för år, 'MS' för månad start, 'AS' för år start, 'B' för business day, 'H' för hour, 'T' för minute, 'S' för second, 'L' för millisecond, 'U' för microsecond, 'N' för nanosecond.

hdisplay([
df.head(10),
df.resample(resample_method).max().fillna(df[["n","rand"]].mean()).head(10)],
["Original", f"Resampled using '{resample_method}'"]
)
```

### Working on real data

```python
apple = pd.read_csv("../data/Historicaldata_Apple.csv", parse_dates=["Date"], index_col="Date")

apple.info()
# ändra date till datetime på date (genom parse_dates)
# Sätta datetimeindex istället för rangeindex
# antingen apple.setindex("Date") eller via parameter index_col
```

```python
# Behöver få bort dollartecken så att de inte ska vara objecttyper
apple.rename(columns={"Close/Last":"Close"}, inplace=True)

apple[['Close', 'Open', 'High', 'Low']] = apple[['Close', 'Open', 'High', 'Low']].map(lambda value: value.replace("$", ""))       #.replace('[\$,]', '', regex=True).astype(float)
                                                                                        # map kör en funktion för varje cell
                                                                                        # Apply gör något på varje element i en serie , alltså hela serien och inte enskilda värdet. 
apple.head()
```

```python
# Konvertera object typ till str

# apple = apple.astype({col: "float64" for col in to_int_columns})

apple[['Close', 'Open', 'High', 'Low']] = apple[['Close', 'Open', 'High', 'Low']].astype("float64")

apple.info()
```

```python
apple.sort_index(inplace=True)    # sort_values för att sortera på en kolumn

apple
```

```python
apple.loc["2018", "Close"].mean()
apple.loc["2018", "High"].max()
```

```python
# pd.set_option("display.float_format", '{:.5f}'.format)         # Man vill använda formateringskod på detta sätt och inte avrunda p.g.a. fortsatt användning och fel som kan uppstå.

pd.set_eng_float_format(accuracy=2, use_eng_prefix=True)

apple.resample("Y").mean()
```
